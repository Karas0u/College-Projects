# Pre-trained Word2Vec Model

The pre-trained Word2vec model by Google is trained on the Google News dataset. 
https://drive.google.com/u/0/uc?id=0B7XkCwpI5KDYNlNUTTlSS21pQmM&export=download.
It has a vocabulary size of 3 million words and phrases and each vector has 300 dimensions.

```python
import gensim
import os
from gensim.models import KeyedVectors
```

### Load pre-trained vectors

```python
import os
path = os.path.join("D:", "IIT MBA", "Sem 4", "NLP", "GoogleNews-vectors-negative300.bin")
model = KeyedVectors.load_word2vec_format(path, binary=True)
```

```python
path = os.path.join("D:", "IIT MBA", "Sem 4", "NLP", "GoogleNews-vectors-negative300.bin")
model = KeyedVectors.load_word2vec_format(path, binary=True)
```

### Finding the word most similar words

```python
model.most_similar('Delhi')
```

```python
model.most_similar('Bangalore', topn=3)
```

### Validating the Relationship : vector(Man) – vector (King) + vector (Queen) = vector (Woman)

```python
result = model.most_similar(positive=['man', 'queen'], negative=['king'], topn=1)
print(result)
```

```python
result = model.most_similar(positive=['man', 'queen'], negative=['king'], topn=2)
print(result)
```

### Validating the Relationship : vector(Spain) + vector(Moscow) - vector(Russia) = vector (Madrid)

```python
result = model.most_similar(positive=['Spain', 'Moscow'], negative=['Russia'], topn=1)
print(result)
```

```python
result = model.most_similar(positive=['Brazil', 'Moscow'], negative=['Russia'], topn=1)
print(result)
```
